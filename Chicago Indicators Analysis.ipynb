Load the SQL magic module


```python
import pandas as pd
census_data = pd.read_csv("ChicagoCensusData.csv")
public_schools = pd.read_csv("ChicagoPublicSchools.csv")
crime_data = pd.read_csv("ChicagoCrimeData.csv")
```

```python
def clean_column_names(dataframe):
    dataframe.columns = dataframe.columns.str.replace('[^0-9a-zA-Z_]', '_', regex=True)  
    dataframe.columns = dataframe.columns.str.replace('__+', '_', regex=True)           
    dataframe.columns = dataframe.columns.str.strip('_')                                
    return dataframe


census_data = clean_column_names(census_data)
public_schools = clean_column_names(public_schools)
crime_data = clean_column_names(crime_data)

print("Census Data Columns:", census_data.columns)
print("Public Schools Columns:", public_schools.columns)
print("Crime Data Columns:", crime_data.columns)

def create_table(cursor, table_name, dataframe):
    columns = ", ".join([
        f"`{col}` TEXT" if dataframe[col].dtype == 'object' else f"`{col}` FLOAT" 
        for col in dataframe.columns
    ])
    create_query = f"CREATE TABLE IF NOT EXISTS {table_name} ({columns})"
    cursor.execute(create_query)

def insert_data(cursor, table_name, dataframe):
    placeholders = ", ".join(["%s"] * len(dataframe.columns))
    insert_query = f"INSERT INTO {table_name} VALUES ({placeholders})"
    for _, row in dataframe.iterrows():
        cursor.execute(insert_query, tuple(row))

create_table(cursor, "ChicagoCensusData", census_data)
insert_data(cursor, "ChicagoCensusData", census_data)

create_table(cursor, "ChicagoPublicSchools", public_schools)
insert_data(cursor, "ChicagoPublicSchools", public_schools)

create_table(cursor, "ChicagoCrimeData", crime_data)
insert_data(cursor, "ChicagoCrimeData", crime_data)

cursor.close()
conn.close()
```

### Problem 1

### Problem 2

##### List community area names and numbers with per capita income less than 11000.


```python
query = "SELECT community_area_name, community_area_number FROM ChicagoCensusData WHERE per_capita_income < 11000;"
cursor.execute(query)
results = cursor.fetchall()
results
```

### Problem 3

##### List all case numbers for crimes involving minors?(children are not considered minors for the purposes of crime analysis) 


```python
query = "SELECT DISTINCT case_number FROM ChicagoCrimeData WHERE description LIKE '%minor%';"
cursor.execute(query)
results = cursor.fetchall()
results
```

### Problem 4

##### List all kidnapping crimes involving a child?


```python
query = "SELECT DISTINCT case_number FROM ChicagoCrimeData WHERE primary_type = 'KIDNAPPING' AND description LIKE '%child%';"
cursor.execute(query)
results = cursor.fetchall()
results
```

### Problem 5

### Problem 8

##### Which community area is most crime prone? Display the coumminty area number only.


```python
query = "SELECT community_area_number FROM ChicagoCrimeData GROUP BY community_area_number ORDER BY COUNT(*) DESC LIMIT 1;"
cursor.execute(query)
result = cursor.fetchone()
result
```

### Problem 9

##### Use a sub-query to find the name of the community area with highest hardship index


```python
query = "SELECT community_area_name,hardship_index FROM ChicagoCensusData WHERE hardship_index = (SELECT MAX(hardship_index) FROM ChicagoCensusData);"
cursor.execute(query)
result = cursor.fetchone()
result
```

### Problem 10

##### Use a sub-query to determine the Community Area Name with most number of crimes?


```python
query = """
SELECT community_area_name
FROM ChicagoCensusData
WHERE community_area_number = (
    SELECT community_area_number
    FROM ChicagoCrimeData
    GROUP BY community_area_number
    ORDER BY COUNT(*) DESC
    LIMIT 1
);
"""
cursor.execute(query)
result = cursor.fetchone()
result
```
